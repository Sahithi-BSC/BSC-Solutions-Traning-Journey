1. What is Natural Language Processing (NLP)?

Answer:

Natural Language Processing (NLP) is a branch of Artificial Intelligence that enables computers to understand, process, analyze, and generate human language. It acts as a bridge between humans and computers by converting human language into a format(numbers 0s & 1s) that computers can understand and process.

2. Why do computers need NLP? Give one real-world example.

Answer:

Computers understand only numbers and machine language, not human languages like English. NLP converts human language into a machine-readable format so computers can understand the meaning.

Example:
Google Translate uses NLP to translate text from one language to another.

3. List four real-life applications of NLP.

Answer:

4. What is a rule-based NLP system?

Answer:

A rule-based NLP system makes decisions using predefined rules instead of machine learning.

Example:
If an email contains the word "Congratulations", the system marks it as Possible Spam.

5. Explain the seven steps of the NLP pipeline.

Answer:

Collect text data
Clean the text
Convert text to lowercase
Tokenization
Remove stopwords
Convert text into numerical features
Train and evaluate the machine learning model

6. What is tokenization?

Answer:

Tokenization is the process of splitting a sentence into smaller units called tokens (usually words).

Example: "I love NLP"
         ["I", "love", "NLP"]

It is necessary because machine learning models process individual words rather than entire sentences.

7. What are stopwords?

Answer:

Stopwords are common words that usually do not add much meaning to a sentence.

Examples:

is
the
and

They are removed to reduce unnecessary data and improve processing speed.

8. Why convert text to lowercase?

Answer:

Lowercasing makes words consistent so the computer treats words with different capitalizations as the same.

Example:

Before:

Apple
apple
APPLE

After:

apple
apple
apple

This avoids counting the same word multiple times.

10. Difference between = and ==.

Answer:

= is used for assigning values.
== is used for comparing values.

Example:

Correct:

df[df["Sentiment"] == "Positive"]

Wrong:

df[df["Sentiment"] = "Positive"]

Using = inside filtering causes a syntax error because Python expects a comparison, not an assignment.

27. Create a movie review dataset and display negative reviews

```python
import pandas as pd

movie_data = {
    "Review": [
        "Amazing movie",
        "Worst movie ever",
        "Excellent acting",
        "Very boring",
        "Loved the story",
        "Waste of time"
    ],
    "Sentiment": [
        "Positive",
        "Negative",
        "Positive",
        "Negative",
        "Positive",
        "Negative"
    ]
}

movies = pd.DataFrame(movie_data)

negative_reviews = movies[movies["Sentiment"] == "Negative"]

print(negative_reviews)
```

28. Find the longest movie review

```python
movies["Length"] = movies["Review"].str.len()

longest_review = movies.loc[movies["Length"].idxmax()]

print(longest_review)
```
