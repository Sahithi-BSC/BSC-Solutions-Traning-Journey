27. Write the Gradient Descent weight update formula.

Answer:

Weight(new) = Weight(old) − (Learning Rate×Gradient)

or

w = w − η (dw/dL)

where:

w = weight
η = learning rate
dw / dL = gradient

50. For a 3-class classification problem with raw output scores [2.0, 1.0, 0.1], use the Softmax formula to calculate the probability of Class 1.

Softmax Formula:

P(xi) =  exi / ∑ex

Given scores:

Class 1 = 2.0
Class 2 = 1.0
Class 3 = 0.1

Calculate exponentials:

e(2.0) = 7.389
e(1.0) = 2.718
e(0.1) = 1.105

Sum:

7.389+2.718+1.105=11.212

Probability of Class 1:

11.212 / 7.389 = 0.659
