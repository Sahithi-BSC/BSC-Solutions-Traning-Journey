```python
def detect_spam(message):
    if "congratulatios" in message or "prize" in message:
        return "Spam "
    else:
        return "Not Spam"

message = input("Enter the Message: ")
print(detect_spam(message))
```

```python
def calculate_accuracy(correct, total):
    accuracy = (correct / total) * 100
    return accuracy

correct = int(input("Enter the number of correct answers: "))

total = int(input("Enter the total number of questions: "))

result = calculate_accuracy(correct, total)

print("Accuracy: ", result, "%")
```

```python
def count_spam(predictions):
    return predictions.count("Spam")

predictions = ["Spam", "Not Spam", "Spam", "Not Spam", "Spam"]

print("total Spam Predictions: ", count_spam(predictions))
```

```python
def store_student(name, age):
    return name,age

def display_student(name, age):
    print("Name: ", name)
    print("Age: ", age)

name = input("Enter the name: ")
age = int(input("Enter the age: "))

student = store_student(name, age)
display_student(student[0], student[1])
```
