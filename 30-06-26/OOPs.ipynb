```python
class Student:
    def __init__(self, name, course):
        self.name = name
        self.course = course

    def display(self):
        print("Name :", self.name)
        print("Course :", self.course)

name = input("Enter name: ")
course = input("Enter course: ")

student = Student(name, course)
student.display()
```

```python
class AIModel:
    def __init__(self, model_name, accuracy, version):
        self.model_name = model_name
        self.accuracy = accuracy
        self.version = version

    def display(self):
        print("Model Name :", self.model_name)
        print("Accuracy :", self.accuracy)
        print("Version :", self.version)

model1 = AIModel("Spam Detector", 0.95, "1.0")
model2 = AIModel("Image Classifier", 0.90, "2.0")

model1.display()
model2.display()
```

```python
class Employee:
    def __init__(self, id, name, salary):
        self.id = id
        self.name = name
        self.salary = salary

    def display(self):
        print("ID :", self.id)
        print("Name :", self.name)
        print("Salary :", self.salary)

emp1 = Employee(1, "John ", 50000)
emp2 = Employee(2, "Jane ", 60000)

emp1.display()
emp2.display()
```
