```python
class car:
    def __init__(self,brand,color):
        self.brand = brand
        self.color = color
    def display(self):
        print("brand:",self.brand)
        print("color:",self.color)
        print("_"*30)

bmw = car("BMW","Black")
audi = car("Audi","White")
tesla = car("Tesla","Red")

bmw.display()
audi.display()
tesla.display()
```

```python
#Encapsulation
class AImodel:
    def __init__(self, model_name):
        self.model_name = model_name
    def display(self):
        print("model name",self.model_name)
model1 = AImodel("Spam Detection Model")
model1.display()
```

```python
#Inheritance
class AImodel:
    def train(self):
        print("Training the model...")
class Spamdetection(AImodel):
    pass
spam = Spamdetection()
spam.train()
```

```python
#Polymorphism
class AImodel:

    def predict(self):
        print("Predicting")

class SpamDetection(AImodel):

    def predict(self):
        print("Predicting spam")

class SentimentAnalysis(AImodel):

    def predict(self):
        print("Predicting sentiment")

spam = SpamDetection()
sentiment = SentimentAnalysis()

spam.predict()
sentiment.predict()
```

```python
#Magic Methods
class student:
    def __init__(self,name,age):
        self.name = name
        self.age = age

student1 = student("John",22)
print(student1.name)
print(student1.age)
```

```python
class Student:
    def __init__(self,name,age):
        self.name = name
        self.age = age

    def __str__(self):
        return f"Student(Name: {self.name}, Age: {self.age})"

student1 = Student("Alice", 21)
print(student1)
```

```python
try:
    num = int(input("Enter Number"))
    result = 100/num
    print("Result:", result)

except ZeroDivisionError:
    print("Division by zero is not allowed.")
except ValueError:
    print("Please enter a valid integer.")

else:
    print("Division completed successfully.")

finally:
    print("Program Execution completed.")
```

```python
#Sentiment Analysis
try:
    Review = input("Enter a Product Review: ")

    if Review.strip() == "":
        raise ValueError("Review cannot be empty.")
    if Review.isdigit():
        raise ValueError("Review should contain only Text. No Number")
    Review = Review.lower()
    positivewords = ["good", "excellent", "amazing", "fantastic", "great"]
    negativewords = ["bad", "poor", "terrible", "awful", "horrible"]
    positivecount = sum(word in Review for word in positivewords)
    negativecount = sum(word in Review for word in negativewords)

    if positivecount > negativecount:
        print("Prediction: Positive Sentiment")
    elif negativecount > positivecount:
        print("Prediction: Negative Sentiment")
    else:
        print("Prediction: Neutral Sentiment")


except ValueError as e:
    print(f"Input Error: {e}")
except Exception as e:
    print(f"An unexpected error occurred: {e}")

else:
    print("Sentiment analysis successfully.")

finally:
    print("AI Prediction completed.")
```
