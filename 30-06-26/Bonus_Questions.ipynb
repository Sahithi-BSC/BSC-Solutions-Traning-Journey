```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def save(self):
        file = open("students.txt", "a")
        file.write(self.name + " " + str(self.age) + "\n")
        file.close()

name = input("Enter name: ")
age = input("Enter age: ")

student = Student(name, age)

student.save()

print("Student information saved.")
```
