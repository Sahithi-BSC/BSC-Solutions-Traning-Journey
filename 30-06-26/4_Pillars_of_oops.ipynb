```python
#Encapsulation
class BankAccount:
    def __init__(self):
        self.__balance = 0

    def deposit(self, amount):
        self.__balance = self.__balance + amount

    def withdraw(self, amount):
        self.__balance = self.__balance - amount

    def check_balance(self):
        return ("Balance: ", self.__balance)

account = BankAccount()

account.deposit(1000)
account.withdraw(500)
account.check_balance()
```

```python
#Inheritance

class AIModel:
    def display(self):
        print("AI Model")

class SpamDetector(AIModel):
    pass

class FaceRecognition(AIModel):
    pass

model1 = SpamDetector()
model2 = FaceRecognition()

model1.display()
model2.display()
```

```python
class SpamDetector:
    def predict(self):
        print("Spam Detected")

class FaceRecognition:
    def predict(self):
        print("Face Recognized")

class ImageClassifier:
    def predict(self):
        print("Image Classified")

model1 = SpamDetector()
model2 = FaceRecognition()
model3 = ImageClassifier()

model1.predict()
model2.predict()
model3.predict()
```

```python
from abc import ABC, abstractmethod

class Model(ABC):
    @abstractmethod
    def train(self):
        pass

class SpamDetector(Model):
    def train(self):
        print("Training Spam Detector")

class FaceRecognition(Model):
    def train(self):
        print("Training Face Recognition")

model1 = SpamDetector()
model2 = FaceRecognition()

model1.train()
model2.train()
```
