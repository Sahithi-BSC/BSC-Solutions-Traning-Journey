```python
name = input("Enter name: ")
age = input("Enter age: ")
course = input("Enter course: ")

file = open("students.txt", "w")

file.write("Name: " + name + "\n")
file.write("Age: " + age + "\n")
file.write("Course: " + course)

file.close()

file = open("students.txt", "r")

print(file.read())

file.close()
```
