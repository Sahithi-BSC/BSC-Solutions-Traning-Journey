41. Perform an Independent t-test between Male and Female Salary.

```python
from scipy import stats

male = data[data["Gender"]=="Male"]["Salary"]
female = data[data["Gender"]=="Female"]["Salary"]

stats.ttest_ind(male, female)
```

42. Find the 95% Confidence Interval for Salary.

```python
mean = data["Salary"].mean()
sem = stats.sem(data["Salary"])

stats.t.interval(0.95, len(data["Salary"])-1, loc=mean, scale=sem)
```

43. Perform a Chi-Square Test between Gender and Left_Company.

```python
table = pd.crosstab(data["Gender"], data["Left_Company"])

stats.chi2_contingency(table)
```

44. Perform One-Way ANOVA on Salary across Departments.

```python
groups = [x["Salary"] for _, x in data.groupby("Department")]

stats.f_oneway(*groups)
```

48. Calculate Cohen's d between Male and Female Salary.

```python
import numpy as np

pooled = np.sqrt((male.var()+female.var())/2)

(male.mean()-female.mean())/pooled
```

49. Perform Bootstrapping (1000 samples) for Salary Mean.

```python
means = []

for i in range(1000):
    sample = data["Salary"].sample(frac=1, replace=True)
    means.append(sample.mean())

np.mean(means)
```

50. Apply Bayes' Theorem to calculate the probability that an employee is in AI given that they have Positive Sentiment.

```python
p_ai = len(data[data["Department"]=="AI"]) / len(data)

p_positive = len(data[data["Sentiment"]=="Positive"]) / len(data)

p_positive_ai = len(data[(data["Department"]=="AI") & (data["Sentiment"]=="Positive")]) / len(data[data["Department"]=="AI"])

(p_positive_ai * p_ai) / p_positive
```
