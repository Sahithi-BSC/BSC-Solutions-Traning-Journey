```python
#Importing the required libraries and loading the dataset

import pandas as pd

df = pd.read_csv('Mall_Customers.csv')
```

```python
# Displaying the first 5 rows of the dataset

df.head(5)
```

```python
# Displaying the shape of the dataset

df.shape
```

```python
# Displaying the information about the dataset

df.info()
```

```python
# Checking for missing values

df.isnull().sum()
```

```python
# Checking for duplicate rows

df.duplicated().sum()
```

```python
# Dropping duplicate rows

df.drop_duplicates(inplace=True)
```

```python
# Displaying the data types of each column

df.dtypes
```

```python
# Displaying the last 5 rows

df.tail()
```

```python
# Renaming the column Genre to Gender

df.rename(columns = {'Genre': 'Gender'}, inplace = True)
```

```python
# Displaying the column names

df.columns
```

```python
# Displaying the number of unique values in each column

df.nunique()
```

```python
# Encoding the categorical variable Gender

from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

df["Gender"] = label_encoder.fit_transform(df["Gender"])
```

```python
# Displaying the first 10 rows after the data cleaning and preproccessing steps

df.head(10)
```

```python
# Displaying the statistical summary of the dataset

df.describe()
```

```python
df.mean(numeric_only=True)
```

```python
df.median(numeric_only=True)
```

```python
# Displaying the count of each unique value in the Gender

df["Gender"].value_counts()
```

```python
# Displaying the correlation matrix of the dataset

df.corr(numeric_only=True)
```

```python
#Histogram - Univariate

import matplotlib.pyplot as plt
plt.hist(df["Age"])
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.show()
```

```python
#Boxplot - Univariate

plt.boxplot(df["Annual Income (k$)"])
plt.title("Annual Income Distribution")
plt.ylabel("Annual Income (k$)")
plt.show()
```

```python
# Annual Income Column

df["Annual Income (k$)"].head(10)
```

```python
#Line plot - Univariate

plt.plot(df["Age"])
plt.title("Age")
plt.xlabel("Index")
plt.ylabel("Age")
plt.show()
```

```python
#Age Column

df["Age"].head(10)
```

```python
#Bar plot - Categorical & Univariate

df["Gender"].value_counts().plot(kind="bar")
plt.title("Gender Distribution")
plt.xlabel("Gender")
plt.ylabel("Count")
plt.show()
```

```python
#Gender Column

df["Gender"].head(10)
```

```python
#Pie chart - Categorical & Univariate

df["Gender"].value_counts().plot(
    kind="pie",
    autopct="%1.1f%%"
)
plt.ylabel("Gender")
plt.show()
```

```python
# Gender Column

df["Gender"].head(10)
```

```python
#Scatter plot - Bivariate

plt.scatter(df["Age"], df["Annual Income (k$)"])
plt.title("Age vs Annual Income")
plt.xlabel("Age")
plt.ylabel("Annual Income (k$)")
plt.show()
```

```python
# Age & Annual Income Columns

df[["Age", "Annual Income (k$)"]].head(10)
```

```python
#Bar Plot - Bivariate

df.groupby("Gender")["Annual Income (k$)"].mean().plot(kind="bar")
plt.title("Average Annual Income by Gender")
plt.xlabel("Gender")
plt.ylabel("Average Annual Income (k$)")
plt.show()
```

```python
# Gender & Annual Income Columns

df[["Gender", "Annual Income (k$)"]].head(10)
```

```python
#Box Plot - Bivariate

df.boxplot(column="Annual Income (k$)", by="Gender")
plt.title("Annual Income by Gender")
plt.xlabel("Gender")
plt.ylabel("Annual Income (k$)")
plt.show()
```

```python
# Gender & Annual Income Columns

df[["Gender", "Annual Income (k$)"]].head(10)
```

```python
#Heatmap - Multivariate

import matplotlib.pyplot as plt

plt.imshow(df.corr(numeric_only=True), cmap="coolwarm")

plt.colorbar()

plt.xticks(range(len(df.corr(numeric_only=True).columns)),
           df.corr(numeric_only=True).columns,
           rotation=45)

plt.yticks(range(len(df.corr(numeric_only=True).columns)),
           df.corr(numeric_only=True).columns)

plt.title("Correlation Heatmap")

plt.show()
```
